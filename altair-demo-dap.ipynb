```python
import altair as alt
from vega_datasets import data

cars = data.cars()
cars.head()
```

```python
alt.Chart(cars).mark_point()
```

```python
alt.Chart(cars).mark_point().encode(x = 'Miles_per_Gallon')
```

```python
alt.Chart(cars).mark_tick().encode(x = 'Miles_per_Gallon')
```

```python
alt.Chart(cars).mark_point().encode(x = 'Miles_per_Gallon',
                                    y = 'Horsepower')
```

```python
alt.Chart(cars).mark_point().encode(x = 'Miles_per_Gallon',
                                    y = 'Horsepower').interactive()
```

```python
alt.Chart(cars).mark_point().encode(x = 'Miles_per_Gallon',
                                    y = 'Horsepower').interactive().to_dict()
```

```python
alt.Chart(cars).mark_bar().encode(x = 'Miles_per_Gallon',
                                    y = 'Horsepower')
```

```python
alt.Chart(cars).mark_bar().encode(x = alt.X('Miles_per_Gallon', bin=True),
                                   y = 'count()')
```

```python
alt.Chart(cars).mark_bar().encode(x = alt.X('Miles_per_Gallon', bin=True),
                                  y = alt.Y('Horsepower', bin=True),
                                  color = 'count()')
```

```python
alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
                                  y = alt.Y('Horsepower'),
                                  color = 'Origin')
```

```python
interval = alt.selection_interval(encodings=['x'])

alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
                                  y = alt.Y('Horsepower'),
                                  color = 'Origin').properties(selection = interval)
```

```python
interval = alt.selection_interval()

alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
            y = alt.Y('Horsepower'),
            color = alt.condition(interval, 'Origin', alt.value('lightgray')))\
.properties(selection = interval)
```

```python
interval = alt.selection_interval()

chart = alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
            y = alt.Y('Horsepower'),
            color = alt.condition(interval, 'Origin', alt.value('lightgray')))\
.properties(selection = interval)

chart | chart
```

```python
interval = alt.selection_interval()

chart = alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
            y = alt.Y('Horsepower'),
            color = alt.condition(interval, 'Origin', alt.value('lightgray')))\
.properties(selection = interval)

chart | chart.encode(x = 'Acceleration')
```

```python
interval = alt.selection_interval(encodings=['x'])

chart = alt.Chart(cars).mark_point().encode(x = alt.X('Miles_per_Gallon'),
            y = alt.Y('Horsepower'),
            color = alt.condition(interval, 'Origin', alt.value('lightgray')))\
.properties(selection = interval)

chart | chart.encode(x = 'Acceleration').save('test.html')
```
